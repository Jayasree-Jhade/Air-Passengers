## method 2 :Augumented Dicky Fuller Test

* for the ADF test
* H0: Data is not stationary 
* H1: Data is stationary
* Based on P-Value , we will accept or reject H0 (if P-Value<5% reject null hypothesis)

```python
from statsmodels.tsa.stattools import adfuller
result=adfuller(df["Passengers"])
result[1]
```

# now we have to convert the non-stationary data to stationary data

```python
# Apply differencing to convert into stationarity
```

```python
diff=df["Passengers"]-df["Passengers"].shift(2)
diff.dropna(inplace=True)

adfuller_result=adfuller(diff)
adfuller_result[1]
```

#### Train-test -split 
* Split the data into train(80%) and test (20%)
* for time series data we have to use first 80% of records for trainig and next 20% for testing
* here 80% = 0.8*142=114 , so first 114 records used for training.

```python
# train test split
```

```python
y_train=diff[:114]
y_test=diff[114:]
```

# Plot ACF/PACF and find p,d,q parameters

```python
from statsmodels.tsa.stattools import acf,pacf
import statsmodels.api as sm
fig=sm.graphics.tsa.plot_acf(diff,lags=40)
fig=sm.graphics.tsa.plot_pacf(diff,lags=40)
```

##### from ACF curve,optimal value of q in the ARIMA model must be 12
##### from PACF curve,optimal value of P in the ARIMA model must be 11

### Build ARIMA model
* ARIMA model is built by using p,d,q values
* p is AR value(determined from PACF plot)
* d is integration
* q is MA value(determined from ACF plot)

```python
# build arima model
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from statsmodels.tsa.arima.model import ARIMA
model=ARIMA(diff,order=(11,0,12))
ARIMA=model.fit()
ypred_train=ARIMA.predict(start=y_train.index[0],end=y_train.index[-1])
ypred_test=ARIMA.predict(start=y_test.index[0],end=y_test.index[-1])

from sklearn.metrics import r2_score
print("train r2:",r2_score(ypred_train,y_train))
print("test r2:",r2_score(ypred_test,y_test))                       
```

### Build SARIMAX model
* whenever there is a seasonility in given data apply SARIMAX.
* SARIMA model is built by using p,d,q,s values
* p is AR value(determined from PACF plot)
* d is integration
* q is MA value(determined from ACF plot
* S is seasonility value

```python
# SARIMAX ( modelling prediction and evaluation)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model=SARIMAX(diff,seasonal_order=(11,0,12,12))
SARIMAX=model.fit()
ypred_train=SARIMAX.predict(start=y_train.index[0],end=y_train.index[-1])
ypred_test=SARIMAX.predict(start=y_test.index[0],end=y_test.index[-1])

from sklearn.metrics import r2_score
print("train r2:",r2_score(ypred_train,y_train))
print("test r2:",r2_score(ypred_test,y_test)) 
```
